# Gradient descent - test

```python
import sys
import random
import numpy as np
import matplotlib.pyplot as plt
```

### Data import

```python
# Physical data
l = 10
tau = 5
dep = 2
# Numerical data
nx = 10
dx = l/nx
nt = 12
dt = tau/nt
x_grid = np.linspace(0,l,nx)
T_operator = l/nx*np.tri(nt, nx, 0, dtype=int)
```

```python
# training data
sigma = 1
mu = l/2
x_dagger = (sigma*np.sqrt(2*np.pi))**-1*np.exp(-(x_grid-mu)**2/2*sigma**2)
x_init = np.zeros(nx)
y = T_operator.dot(x_dagger)
```

```python
plt.plot(np.linspace(0,1,nx),x_dagger)
```

```python
plt.plot(np.linspace(0,1,nt),y)
```

```python
### Gradient descent
```

```python
nstep = 20000
precision = 0.01
coeff = 0.02

error = np.zeros(nstep)
xk = np.zeros((nx,nstep))
xk[:,0] = x_init.copy()

for k in range(nstep-1):
    xk[:,k+1] = xk[:,k] \
    - coeff*(np.dot(T_operator.T,T_operator).dot(xk[:,k]) \
             - np.dot(T_operator.T,y))
    error[k] = np.linalg.norm(x_dagger - xk[:,k])
    if error[k]<precision :
        print("objectif atteint en {} itérations".format(k))
    
```

```python
plt.plot(np.linspace(0,1,nx), xk[:,nstep-1],'+')
plt.plot(np.linspace(0,1,nx),x_dagger)
```
