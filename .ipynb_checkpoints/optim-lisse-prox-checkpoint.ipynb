# Algorithme d'optimisation pour problème lisse

Algorithme d'optimisation d'un problème lisse :
 - Descente de gradient (ajouter des termes d'accélérations à la Nesterov),
 - Newton,
 - Quasi-Newton, 
 - Forward-backward, 
 - Douglas-Rachford, 
 - Chambolle-pock.
 - (Avec pénalisation)Frank-Wolfe.


## Definition du problème inverse

Soit un vecteur $y =(y_j) \in Y = \mathbb{R}^{nt}$,
on cherche à optimiser la fonction sur $x =(x_i) \in X = \mathbb{R}^{nx}$ avec $\alpha >0$

$$
\text{minimize} \; \frac{1}{2}\| Tx-y\|_Y^2 + \alpha\frac{1}{2} \|x\|_X^2
$$

avec

$$
T x = \sum_{i=0}^{nx-1} x_i
$$

$$
\| y \|_Y^2 = \sum_{j=0}^{nt} y_j^2 \frac{\tau}{nt}
$$


$$
\| x \|_X^2 = \sum_{j=0}^{nx} x_i^2 \frac{\ell}{nx}
$$

```python
l = 3
v = 2
tau = 2.5
nx = 200
nt = 350
grid = np.linspace(0,l,nx,endpoint=False)
```

```python
cfl = v*(tau/nt)/(l/nx)
print("La condition cfl est {}".format(cfl))
flow = (1-cfl)*np.eye(nx) + cfl*np.diag(np.ones(nx-1),-1)

# moment of order k
xk = l/nx*np.ones(nx)
T = np.zeros((nt,nx))
for i in range(nt):
    T[i,:] = xk
    xk = flow.dot(xk)
```

```python
mu, sigma = l/2, 0.1*l # mean and standard deviation
xstar = 1/(sigma*math.sqrt(2*math.pi))*np.exp(-0.5*(grid-mu)**2/sigma**2)
xstarp = 1/(sigma*math.sqrt(2*math.pi))*np.exp(-0.5*(grid-mu)**2/sigma**2)*(-(grid-mu)/sigma**2)
noise = 0.1
xi = noise*np.random.normal(0,0.1, nt)
y = T.dot(xstar)
yd = T.dot(xstar) + xi
```

```python
# taille de la figure
plt.figure(figsize=(8,5))
grid_t = np.linspace(0,tau,nt)
plt.title("Distribution initiale $x$")
plt.plot(grid, xstar, label="solution exacte")
plt.legend()
plt.show()
```

```python
# taille de la figure
plt.figure(figsize=(8,5))
grid_t = np.linspace(0,tau,nt)
plt.title("Moment d'ordre 0, $y=Tx$ et $y^\delta$")
plt.plot(grid_t, y, label="mesure")
plt.plot(grid_t, yd, label="mesure bruitée")
plt.legend()
plt.show()
```

```python
np.linalg.norm(y-yd)/np.linalg.norm(y)
```

```python
normH1 = l/nx*np.linalg.norm(xstar)+l/nx*np.linalg.norm(xstarp)
alpha = math.sqrt(tau/nt*np.linalg.norm(y-yd))/math.sqrt(normH1)
alpha
```

```python
from scipy.linalg import eigvalsh 
L = eigvalsh(tau/nt*np.transpose(T).dot(T)).max()
L
```

```python
def f(x):
    return 0.5*tau/nt*np.linalg.norm(T.dot(x)-yd)**2+0.5*alpha*l/nx*np.linalg.norm(x)**2
def df(x):
    return tau/nt*np.transpose(T).dot(T.dot(x)-yd)+alpha*l/nx*x
def ddf(x):
    return tau/nt*np.transpose(T).dot(T) + alpha*l/nx*np.eye(nx)
```

# Descente de gradient 

```python
def Descente_gradient(x0,f,df,tau=1/(2*L),eps=10**-6,Niter=100000):
    xk = x0.copy()
    for k in range(Niter):
        xk-= tau*df(xk)
        if np.linalg.norm(df(xk))<eps:
            print("Number of iterations {}".format(k))
            return xk
    print("Final error eps = {}".format(np.abs(f(xk))))
    print("Max iteration of {} reached".format(Niter))
    return xk  
```

```python
x0 = np.transpose(T).dot(yd)
xhat0 = Descente_gradient(x0,f,df)
```

```python
# taille de la figure
plt.figure(figsize=(8,5))
plt.title("Processus de Nesterov")
plt.plot(grid, xstar, label="solution exacte")
plt.plot(grid, xhat0, label="solution optimisée")
plt.legend()
plt.show()
```

# Processus de Nesterov

```python
def Nesterov(x0,f,df,tau=1/(2*L),eps=10**-16,Niter=10000):
    xk = x0.copy()
    xkm = x0.copy()
    for k in range(Niter):
        yk = xk+(k/2)/(1+(k+1)/2)*(xk - xkm)
        xk, xkm = yk-tau*df(yk), xk
        if np.linalg.norm(df(xk))<eps:
            print("Number of iterations {}".format(k))
            return xk
    print("Final error eps = {}".format(np.abs(f(xk))))
    print("Max iteration of {} reached".format(Niter))
    return xk  
```

```python
x0 = np.transpose(T).dot(yd)
xhat1 = Nesterov(x0,f,df)
```

```python
# taille de la figure
plt.figure(figsize=(8,5))
plt.title("Processus de Nesterov")
plt.plot(grid, xstar, label="solution exacte")
plt.plot(grid, xhat1, label="solution optimisée")
plt.legend()
plt.show()
```

## Newton

```python
def Newton(x0,f,df,ddf,eps=10**-16,Niter=10000):
    xk = x0.copy()
    for k in range(Niter):
        ddfk = ddf(xk)
        if np.linalg.norm(df(xk))<eps:
            print("Number of iterations {}".format(k))
            return xk
        if np.linalg.norm(ddfk)<eps:
            raise Exception(f"La dérivée |ddf| = {np.linalg.norm(ddfk)} est trop petite")
        ddf_inv = np.linalg.inv(ddfk)
        xk-= ddf_inv.dot(df(xk))
    print("Final error eps = {}".format(np.abs(f(xk))))
    print("Max iteration of {} reached".format(Niter))
    return xk
```

```python
x0 = np.transpose(T).dot(yd)
xhat2 = Newton(x0,f,df,ddf)
```

```python
# taille de la figure
plt.figure(figsize=(8,5))
plt.title("Newton iterations")
plt.plot(grid, xstar, label="solution exacte")
plt.plot(grid, xhat2,'g+',label="solution optimisée")
plt.legend()
plt.show()
```

Conclusion : pour obtenur des résultats théoriques, il faut prendre un problème très spécifique, mais qui rentre à peu près dans les cases de l'analyse convexe, et en déduire des propriétés supplémentaires/des améliorations algorithmiques, qui peuvent être validées ou non expérimentalement.

## Quasi-Newton - BFGS method

Given starting point x0, convergence tolerance

Inverse Hessian approximation H0; 

k ← 0;

while ∇fk > eps ;

Compute search direction
pk = -Hk $\nabla$ fk > 0,

Set xk+1=xk + αk pk where αk is computed from a line search procedure to satisfy the Wolfe conditions (3.6);
he values c1 =10−4 and c2 = 0.9 are commonly used

Define sk=xk+1 −xk 
and yk=∇fk+1 −∇fk; 

Compute Hk+1 by means of (6.17);

k ← k + 1;

end (while)

```python
def BFGS(x0,f,df,eps=10**-12,Niter=10):
    print(eps)
    c1, c2 = 10**-4, 0.9
    xk = x0.copy()
    dfk = df(xk)
    Hk = np.linalg.inv(ddf(xk))
    for k in range(Niter):
        if np.linalg.norm(dfk)<eps:
            print("Number of iterations {}".format(k))
            return xk
        # pk
        pk = - Hk.dot(dfk)
        # alphak
        alphak = 1
        # Wolfe condition
        cond1 = f(xk+alphak*xk)-f(xk)- c1*alphak*dfk.T.dot(pk)
        cond2 = df(xk+alphak*xk).T.dot(pk)-c2*dfk.T.dot(pk)
        i=0
        while (cond1>0) or (cond2>0) :
            alphak-=0.01
            # Wolfe condition
            cond1 = f(xk+alphak*xk)-f(xk)- c1*alphak*dfk.T.dot(pk)
            cond2 = df(xk+alphak*xk).T.dot(pk)-c2*dfk.T.dot(pk)
            if i>1000:
                return 0
            i+=1
        # xk+1
        xkm = xk.copy()
        xk = xkm + alphak*pk 
        # sk, yk, rhok
        sk = np.array([xk - xkm])
        yk = np.array([df(xk)- dfk])
        rhok = (yk.dot(sk.T))**-1
        # Hk
        Hk = (np.eye(nx)-rhok*sk.T.dot(yk)).dot(Hk).dot(np.eye(nx)-rhok*yk.T.dot(sk)) + rhok*sk*sk.T
        # update
        dfk = df(xk)
    print("Final error eps = {}".format(np.linalg.norm(dfk)))
    print("Max iteration of {} reached".format(Niter))
    return xk
```

```python
x0 = np.transpose(T).dot(yd)
xhat3 = BFGS(x0,f,df)
```

```python
plt.figure(figsize=(8,5))
plt.title("BFGS iterations")
plt.plot(grid, xstar, label="solution exacte")
plt.plot(grid, xhat3,'g+',label="solution optimisée")
plt.legend()
plt.show()
```

## Forward-backward

#### Explicit Forward-backward

```python
def Explicit_FB(x0,f,g,df,dg,tau= 1/(2*L),eps=10**-12,Niter=10000):
    xk = x0.copy()
    yk = x0.copy()
    for k in range(Niter):
        yk = xk - tau*df(xk)
        xk = yk - tau*dg(yk)
        if np.linalg.norm(df(xk)+dg(xk))<eps:
            print("Number of iterations {}".format(k))
            return xk
    print("Final error eps = {}".format(np.abs(f(xk))))
    print("Max iteration of {} reached".format(Niter))
    return xk  
```

```python
def f(x):
    return 0.5*tau/nt*np.linalg.norm(T.dot(x)-yd)**2
def df(x):
    return tau/nt*np.transpose(T).dot(T.dot(x)-yd)
def g(x):
    return 0.5*alpha*l/nx*np.linalg.norm(x)**2
def dg(x):
    return alpha*l/nx*x
```

```python
x0 = np.transpose(T).dot(yd)
xhat4 = Explicit_FB(x0,f,g,df,dg)
```

```python
plt.figure(figsize=(8,5))
plt.title("Explicit Forward-Backward iterations")
plt.plot(grid, xstar, label="solution exacte")
plt.plot(grid, xhat4,'g+',label="solution optimisée")
plt.legend()
plt.show()
```

#### Implicit Forward-backward

Option 1 :
Compute explicitly the proximal operator.
$$
f(x) = \frac{1}{2} \| Tx-y\|_Y^2
$$

$$
g(x) = \frac{1}{2} \alpha \| x\|_X^2
$$

$$ 
prox_{\gamma g} = \frac{x}{\gamma \alpha+1}
$$

```python
def proxg(x,gamma):
    return x/(gamma*alpha+1)
```

```python
def Forward_backward(x0,df,proxg,tau= 1/(2*L),eps=10**-12,Niter=100000):
    xk = x0.copy()
    yk = x0.copy()
    for k in range(Niter):
        yk = xk - tau*df(xk)
        xk = proxg(yk,tau)
        if np.linalg.norm(df(xk))<eps:
            print("Number of iterations {}".format(k))
            return xk
    print("Final error eps = {}".format(np.abs(f(xk))))
    print("Max iteration of {} reached".format(Niter))
    return xk  
```

```python
x0 = np.transpose(T).dot(yd)
xhat5 = Forward_backward(x0,df,proxg)
```

```python
plt.figure(figsize=(8,5))
plt.title("Explicit Forward-Backward iterations")
plt.plot(grid, xstar, label="solution exacte")
plt.plot(grid, xhat5,'g+',label="solution optimisée")
plt.legend()
plt.show()
```
