```python
# general
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.linalg import inv,pinvh,eig,eigh
from scipy.stats import linregress
%matplotlib inline
import numpy.polynomial.chebyshev as che
# local
from FBResNet.myfunc import Physics
```

```python
a     = 1/2
noise = 0.05
```

```python
# Matrice op√©rateur
nx  = 2000
dx  = 1/nx
m   = 500
T   = dx*np.tri(nx)
```

```python
Ta = np.zeros((nx,nx))
for i in range(nx):
    for j in range(nx):
        if j<=i:
            Ta[i,j]=1/a*dx**(a)*((i-j+1)**a-(i-j)**a)
```

```python
t   = np.linspace(0,1-1/nx,nx)
x   = np.exp(-(t-0.5)**2/0.1**2)
x   = x/np.amax(x)
y   = Ta.dot(x)
```

```python
# noise
m = 500
physics     = Physics(nx,m)
vn          = np.zeros(m)
fmax        = 400
vn[fmax:]   = np.random.randn(m-fmax)
vn_elt      = physics.BasisChangeInv(vn)
vn_elt      = vn_elt/np.linalg.norm(vn_elt)
yd          = y + noise*np.linalg.norm(y)*vn_elt
```

```python
# cheb
nc   = 50
yd_che = che.chebfit(t,yd,nc-1)
x_che = che.chebfit(t,x,nc-1)
pol = np.zeros((nc,nc))
fac = 1/a
for j in range(nc):
    pol[:,j]=fac*che.chebfit(t,t**(a+i),nc-1)
    fac*=(j+1)/(a+j+1)
```

```python
plt.plot(yd_che)
plt.plot(pol.dot(x_che))
```
