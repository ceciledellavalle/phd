```python
import torch
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from FBResNet.pretreat import CreateDataSet
from FBResNet.myfunc import Physics
```

```python
a    = 1
p    = 1
nx   = 2000
m    = 50
test = Physics(nx,m,a,p)
path = './Datasets'
```

```python
lisse, lisse_cos, blurred, blurred_cos =CreateDataSet(test,
                                                      path,
                                                      noise=0.1,
                                                      nsample=5,
                                                      save='no')
```

```python
plt.plot(lisse[0])
plt.plot(blurred[0])
```

```python
tTy  = test.ComputeAdjoint(test.Compute(lisse[0]))
tTye =blurred_cos[0]
plt.plot(tTy)
plt.plot(tTye)
print("error =",np.linalg.norm(tTy-tTye))
```
