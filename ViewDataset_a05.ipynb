```python
from scipy.interpolate import interp1d
# interpolation
def reduce(f,n1,n2):
    t      = np.linspace(0,1,n1, endpoint=True)
    te     = np.linspace(0,1,n2,endpoint=True)
    g      = interp1d(t,f)
    f_red  = g(te)
    return f_red
```
